```python
#1
import pandas as pd

data = {'FirstName': ['Alice', 'Bob'], 'LastName': ['Cruz', 'Dela Rosa']}
df = pd.DataFrame(data)
df['FullName'] = df['FirstName'] + ' ' + df['LastName']
print(df[['FullName']])
```

```python
#2
data = {'Name': ['Alice', 'Bob', 'Cara'], 'Age': [25, None, 30]}
df = pd.DataFrame(data)
df['Age'] = df['Age'].fillna(0)
print(df)
```

```python
#3
data = {'FirstName': ['Juan'], 'MiddleName': ['Santos'], 'LastName': ['Cruz']}
df = pd.DataFrame(data)
df = df.drop('MiddleName', axis=1)
print(df)
```

```python
#4
data = {'Name': ['Alice'], 'Score': [90]}
df = pd.DataFrame(data)
df = df.rename(columns={'Name': 'StudentName', 'Score': 'ExamScore'})
print(df)
```

```python
#5
data = {'Name': ['Alice', 'Bob'], 'Score': [88, 95]}
df = pd.DataFrame(data)
df = df.sort_values(by='Score', ascending=False)
print(df)
```

```python
#6
data = {'Employee': ['Alice', 'Bob', 'Cara'], 'Department': ['IT', 'HR', 'IT']}
df = pd.DataFrame(data)
print(df['Department'].nunique())
```

```python
#7
data = {'Name': ['Alice', 'Bob'], 'Score': [80, 70]}
df = pd.DataFrame(data)
df['Passed'] = df['Score'] >= 75
print(df[['Name', 'Passed']])
```

```python
#8
data = {'Name': ['Alice', 'Bob'], 'Salary': [45000, 55000]}
df = pd.DataFrame(data)
df = df[df['Salary'] >= 50000]
print(df)
```

```python
#9
data = {'Name': ['Alice', 'Bob'], 'Gender': ['F', 'M']}
df = pd.DataFrame(data)
df['Gender'] = df['Gender'].replace({'M': 'Male', 'F': 'Female'})
print(df)
```

```python
#10
data = {'Name': ['Alice'], 'Height_m': [1.6], 'Weight_kg': [55]}
df = pd.DataFrame(data)
df['BMI'] = df['Weight_kg'] / (df['Height_m'] ** 2)
df['BMI'] = df['BMI'].round(2)
print(df[['Name', 'BMI']])
```

```python
#11
data = {'Name': ['A', 'B', 'C'], 'Department': ['HR', 'IT', 'HR'], 'Salary': [40000, 60000, 50000]}
df = pd.DataFrame(data)
result = df.groupby('Department')['Salary'].mean().reset_index()
print(result)
```

```python
#12
customers = {'CustomerID': [1, 2], 'Name': ['Alice', 'Bob']}
orders = {'OrderID': [101], 'CustomerID': [1], 'Total': [500]}
df_customers = pd.DataFrame(customers)
df_orders = pd.DataFrame(orders)
result = pd.merge(df_orders, df_customers, on='CustomerID')[['OrderID', 'Name', 'Total']]
print(result)
```

```python
#13
data = {'Region': ['North', 'South', 'North'], 'Product': ['A', 'A', 'B'], 'Sales': [100, 150, 200]}
df = pd.DataFrame(data)
pivot = df.pivot_table(index='Product', columns='Region', values='Sales', aggfunc='sum', fill_value=0).reset_index()
print(pivot)
```

```python
#14
data = {'Date': ['2023-01-01', '2023-01-02'], 'Sales': [100, 150]}
df = pd.DataFrame(data)
df['CumulativeSales'] = df['Sales'].cumsum()
print(df[['Date', 'CumulativeSales']])
```

```python
#15
data = {'Date': ['2023-01-15']}
df = pd.DataFrame(data)
df['Month'] = pd.to_datetime(df['Date']).dt.strftime('%B')
print(df)
```

```python
#16
data = {'Name': ['Alice', 'Bob'], 'Age': [25, None], 'Salary': [None, 30000]}
df = pd.DataFrame(data)
result = df.isnull().sum().reset_index()
result.columns = ['Column', 'Nulls']
print(result)
```

```python
#17
data = {'Name': ['Alice', 'Bob', 'Cara'], 'Department': ['HR', 'HR', 'IT'], 'Salary': [50000, 60000, 55000]}
df = pd.DataFrame(data)
result = df.sort_values(['Department', 'Salary'], ascending=[True, False]).groupby('Department').head(1)
print(result)
```

```python
#18
data = {'StartDate': ['2023-01-01'], 'EndDate': ['2023-01-10']}
df = pd.DataFrame(data)
df['DaysActive'] = (pd.to_datetime(df['EndDate']) - pd.to_datetime(df['StartDate'])).dt.days
print(df[['DaysActive']])
```

```python
#19
data = {'Email': ['user1@gmail.com', 'user2@yahoo.com']}
df = pd.DataFrame(data)
df = df[df['Email'].str.contains('@gmail.com')]
print(df)
```

```python
#20
data = {'Score': [80, 90, 100]}
df = pd.DataFrame(data)
df['ZScore'] = ((df['Score'] - df['Score'].mean()) / df['Score'].std()).round(1)
print(df[['ZScore']])
```

```python
#21
data = {'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04'], 'Sales': [100, 200, 300, 400]}
df = pd.DataFrame(data)
df['RollingAvg'] = df['Sales'].rolling(window=3).mean()
df = df.dropna()[['Date', 'RollingAvg']]
print(df)
```

```python
#22
data = {'Region': ['North', 'North', 'South'], 'Product': ['A', 'B', 'A'], 'Sales': [100, 150, 200]}
df = pd.DataFrame(data)
result = df.groupby(['Region', 'Product'])['Sales'].sum().reset_index(name='TotalSales')
print(result)
```

```python
#23
data = {'Date': ['2023-01-01', '2023-01-15', '2023-02-01'], 'Sales': [100, 200, 300]}
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.strftime('%b')
result = df.groupby('Month')['Sales'].sum().reset_index(name='TotalSales')
print(result)
```

```python
#24
data = {'Score': [70, 75, 85, 100, 150]}
df = pd.DataFrame(data)
q1 = df['Score'].quantile(0.25)
q3 = df['Score'].quantile(0.75)
iqr = q3 - q1
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr
df['IsOutlier'] = (df['Score'] < lower) | (df['Score'] > upper)
print(df)
```

```python
#25
data = {'Product': ['A', 'B', 'C'], 'Color': ['Red', 'Blue', 'Green']}
df = pd.DataFrame(data)
encoded = pd.get_dummies(df['Color'], prefix='Color')
df = pd.concat([df[['Product']], encoded], axis=1)
print(df.to_string(index=False))
```
