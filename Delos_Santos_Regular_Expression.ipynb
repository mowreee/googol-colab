```python
#1
import re

text = "I have an apple and an orange."
pattern = r"apple"
matches = re.findall(pattern, text)
print(matches)
```

```python
#2
import re

text = "My phone number is 123-456-7890."
pattern = r"\d"
matches = re.findall(pattern, text)
print(matches)
```

```python
#3
import re

text = "I have 2 apples and 10 oranges."
pattern = r"\b\d+\b"
matches = re.findall(pattern, text)
print(matches)
```

```python
#4
import re

text = "The price is 12.50 dollars and discount is 0.75."
pattern = r"\b\d+\.\d+\b"
matches = re.findall(pattern, text)
print(matches)
```

```python
#5
import re

text = "Contact us at support@example.com or sales@example.org."
pattern = r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}"
matches = re.findall(pattern, text)
print(matches)
```

```python
#6
import re

text = "Call me at 123-456-7890 or 987-654-3210."
pattern = r"\b\d{3}-\d{3}-\d{4}\b"
matches = re.findall(pattern, text)
print(matches)
```

```python
#7
import re

text = "Events on 12/05/2023 and 01/01/2024."
pattern = r"\b\d{2}/\d{2}/\d{4}\b"
matches = re.findall(pattern, text)
print(matches)
```

```python
#8
import re

text = "Visit https://example.com or http://test.org for info."
pattern = r"https?://[A-Za-z0-9._%+-]+\.[A-Za-z]{2,}"
matches = re.findall(pattern, text)
print(matches)
```

```python
#9
import re

text = "Alice and Bob went to New York City."
pattern = r"\b[A-Z][a-zA-Z]*\b"
matches = re.findall(pattern, text)
print(matches)
```

```python
#10
import re

text = "Loving the #sunshine and #beach vibes!"
pattern = r"#\w+"
matches = re.findall(pattern, text)
print(matches)
```

```python
#11
import re

text = "Send mail to 12345 or 12345-6789."
pattern = r"\b\d{5}(?:-\d{4})?\b"
matches = re.findall(pattern, text)
print(matches)
```

```python
#12
import re

text = "This is a test with some four letter words like code and play."
pattern = r"\b\w{4}\b"
matches = re.findall(pattern, text)
print(matches)
```

```python
#13
import re

text = "Remove all \t whitespace\n characters."
pattern = r"\s+"
result = re.sub(pattern, "", text)
print(result)
```

```python
#14
import re

text = "<div><p>Hello</p></div>"
pattern = r"<[^>]+>"
matches = re.findall(pattern, text)
print(matches)
```

```python
#15
import re

text = "The function(x) returns (y + z)."
pattern = r"\((.*?)\)"
matches = re.findall(pattern, text)
print(matches)
```

```python
#16
import re

text = "Localhost IP is 127.0.0.1 and router IP is 192.168.1.1."
pattern = r"\b(?:\d{1,3}\.){3}\d{1,3}\b"
matches = re.findall(pattern, text)
print(matches)
```

```python
#17
import re

text = "This  sentence   has    too many     spaces."
pattern = r"\s+"
result = re.sub(pattern, " ", text)
print(result)
```

```python
#18
import re

text = "I am running and swimming in the morning."
pattern = r"\b\w+ing\b"
matches = re.findall(pattern, text)
print(matches)
```

```python
#19
import re

text = "Meeting at 09:30 and 14:45 today."
pattern = r"\b([01]\d|2[0-3]):[0-5]\d\b"
matches = re.findall(pattern, text)
print(matches)
```

```python
#20
import re

text = "Passwords: Passw0rd!, password, P@ss1."
pattern = r"\b(?=.*[A-Z])(?=.*\d)(?=.*[^A-Za-z0-9]).+\b"
matches = re.findall(pattern, text)
print(matches)
```

```python
#21
import re

text = "My gym's rhythm is strong."
pattern = r"\b[^aeiouAEIOU\s]+\b"
matches = re.findall(pattern, text)
print(matches)
```

```python
#22
import re

text = 'She said, "Hello, world!" and then left.'
pattern = r'"(.*?)"'
matches = re.findall(pattern, text)
print(matches)
```

```python
#23
import re

text = "This is is a test test sentence."
pattern = r"\b(\w+)\s+\1\b"
matches = re.findall(pattern, text)
print([f"{match} {match}" for match in matches])
```

```python
#24
import re

text = "Colors used: #FFA07A, #000, and #123ABC."
pattern = r"#(?:[A-Fa-f0-9]{3}|[A-Fa-f0-9]{6})\b"
matches = re.findall(pattern, text)
print(matches)
```

```python
#25
import re

text = "Cards: 4111111111111111, 5500000000000004, 340000000000009."
pattern = r"\b(?:4\d{12}(?:\d{3})?|5[1-5]\d{14})\b"
matches = re.findall(pattern, text)
print(matches)
```
