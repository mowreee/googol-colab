```python
#1
import re

text = "I have an apple and an orange."
result = re.findall(r"apple", text)
print(result)
```

```python
#2
import re

text = "My phone number is 123-456-7890."
result = re.findall(r"\d", text)
print(result)
```

```python
#3
import re

text = "I have 2 apples and 10 oranges."
result = re.findall(r"\d+", text)
print(result)
```

```python
#4
import re

text = "The price is 12.50 dollars and discount is 0.75."
result = re.findall(r"\d+\.\d+", text)
print(result)
```

```python
#5
import re

text = "Contact us at support@example.com or sales@example.org."
result = re.findall(r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}", text)
print(result)
```

```python
#6
import re

text = "Call me at 123-456-7890 or 987-654-3210."
result = re.findall(r"\d{3}-\d{3}-\d{4}", text)
print(result)
```

```python
#7
import re

text = "Events on 12/05/2023 and 01/01/2024."
result = re.findall(r"\d{2}/\d{2}/\d{4}", text)
print(result)
```

```python
#8
import re

text = "Visit https://example.com or http://test.org for info."
result = re.findall(r"https?://[A-Za-z0-9.-]+\.[A-Za-z]{2,}", text)
print(result)
```

```python
#9
import re

text = "Alice and Bob went to New York City."
result = re.findall(r"\b[A-Z][a-z]*\b", text)
print(result)
```

```python
#10
import re

text = "Loving the #sunshine and #beach vibes!"
result = re.findall(r"#\w+", text)
print(result)
```

```python
#11
import re

text = "Send mail to 12345 or 12345-6789."
result = re.findall(r"\b\d{5}(?:-\d{4})?\b", text)
print(result)
```

```python
#12
import re

text = "This is a test with some four letter words like code and play."
result = re.findall(r"\b[a-zA-Z]{4}\b", text)
print(result)
```

```python
#13
import re

text = "Remove all \t whitespace\n characters."
result = re.sub(r"\s+", "", text)
print(result)
```

```python
#14
import re

text = "<div><p>Hello</p></div>"
result = re.findall(r"<[^>]+>", text)
print(result)
```

```python
#15
import re

text = "The function(x) returns (y + z)."
result = re.findall(r"\(([^)]+)\)", text)
print(result)
```

```python
#16
import re

text = "Localhost IP is 127.0.0.1 and router IP is 192.168.1.1."
result = re.findall(r"\b(?:\d{1,3}\.){3}\d{1,3}\b", text)
print(result)
```

```python
#17
import re

text = "This   sentence  has   too many    spaces."
result = re.sub(r"\s+", " ", text)
print(result)
```

```python
#18
import re

text = "I am running and swimming in the morning."
result = re.findall(r"\b\w+ing\b", text)
print(result)
```

```python
#19
import re

text = "Meeting at 09:30 and 14:45 today."
result = re.findall(r"\b\d{2}:\d{2}\b", text)
print(result)
```

```python
#20
import re

text = "Passwords: Passw0rd!, password, P@ss1."
result = re.findall(r"\b(?=.*[A-Z])(?=.*\d)(?=.*[!@#$%^&*()_+{}\[\]:;<>,.?/~`-]).+\b", text)
print(result)
```

```python
#21
import re

text = "My gym's rhythm is strong."
result = re.findall(r"\b[^aeiouAEIOU\s]+\b", text)
print(result)
```

```python
#22
import re

text = 'She said, "Hello, world!" and then left.'
result = re.findall(r'"(.*?)"', text)
print(result)
```

```python
#23
import re

text = "This is is a test test sentence."
result = re.findall(r"\b(\w+)\s+\1\b", text)
print(result)
```

```python
#24
import re

text = "Colors used: #FFA07A, #000, and #123ABC."
result = re.findall(r"#[0-9A-Fa-f]{3,6}\b", text)
print(result)
```

```python
#25
import re

text = "Cards: 4111111111111111, 5500000000000004, 340000000000009."
result = re.findall(r"\b(?:4\d{12}(?:\d{3})?|5[1-5]\d{14})\b", text)
print(result)
```
