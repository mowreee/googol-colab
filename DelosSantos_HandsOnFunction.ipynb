```python
#1
def pat1(num):
    for i in range(num):
        print ('* ' * num)
#2
def pat2(num):
  for i in range(1, num + 1):
    for j in range(i):
        print("*", end=" ")
    print()
#3
def pat3(num):
    for i in range(num,0,-1):
        for j in range(i):
            print("*", end=" ")
        print()
#4
def pat4(num):
    for i in range(num):
        print(' '*(num-i-1) + '* '*(i+1) )
    for i in range(num):
        print(' '*(i+1) + '* '*(num-i-1))
#5
def pat5(num):
  for i in range(num):
    for j in range(num):
        if i == 0 or i == num - 1 or j == 0 or j == num - 1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
#6
def pat6(num):
  for i in range(1, num + 1):
    for j in range(i):
        print("*", end=" ")
    for k in range(2 * (num - i)):
        print(" ", end=" ")
    for j in range(i):
        print("*", end=" ")
    print()

  for i in range(num - 1, 0, -1):
    for j in range(i):
        print("*", end=" ")
    for k in range(2 * (num - i)):
        print(" ", end=" ")
    for j in range(i):
        print("*", end=" ")
    print()

while True:
    num = int(input("Enter number: "))
    print("( 1 - 2 - 3 - 4 - 5 - 6 ) ")
    pat = int(input("Enter type of pattern : "))

    if pat == 1:
        pat1(num)
    elif pat == 2:
        pat2(num)
    elif pat == 3:
        pat3(num)
    elif pat == 4:
        pat4(num)
    elif pat == 5:
        pat5(num)
    elif pat == 6:
        pat6(num)
    else:
        print("Invalid choice")

    choice = input("Do you want to try again? (y/n): ").lower()

    if choice == 'n':
        print("Program closed...")
        break
    elif choice != 'y':
        print("Invalid input. Exiting.")
        break

#No AI
```
